# scikit-learn을 활용한 기후 변화 예측 모델

```python
#1. numpy로 데이터 불러오기

import numpy as np

year, temp_diff = np.loadtxt('5-year-mean-1882-2014.csv', delimiter = ',', usecols = (0, 1), unpack = True)
```

```python
#2. data 살펴보기

print(year[:9])
print(temp_diff[:9])
```

```python
#3. data 시각화하기

import matplotlib.pyplot as plt
plt.scatter(year, temp_diff)
plt.title("scatter plot of temp diff vs year")
plt.xlabel('year')
plt.ylabel('temp_diff')
plt.show()
```

```python
#4. Regression model

from sklearn.linear_model import LinearRegression

linear_model = LinearRegression(fit_intercept = True) #fit_intercept가 false면 ax + b에서 b를 학습하지 않음

year = year.reshape(133, 1) #학습을 위해 하나의 어레이에 담긴 데이터를 하나하나 따로 묶어줌
linear_model.fit(year, temp_diff)
```

```python
#5. model 예측

model_predict = linear_model.predict(year)
print(model_predict[0:9])
```

```python
#6. model 시각화

plt.scatter(year, temp_diff)
plt.plot(year, model_predict)
plt.title('real temp diff vs model predicted temp diff')
plt.show()
```

```python
#7. model 확인
print('y = {0} * x + {1}'.format(linear_model.coef_[0], linear_model.intercept_))
```
